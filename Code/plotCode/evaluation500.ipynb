```python
import matplotlib.pyplot as plt
import numpy as np

# Metrics and scores
metrics = ['Similarity', 'BERTScore', 'BLEURT']
t5_scores = [0.64706, 0.8644, -0.95572]
llama_scores = [0.58832, 0.84866, -0.52462]

x = np.arange(len(metrics))
width = 0.35

# Set style and colors
plt.style.use('classic')
colors = ['#36B37E', '#2684FF']  # Softer green and blue

fig, ax = plt.subplots(figsize=(10, 6))
fig.patch.set_facecolor('white')
ax.set_facecolor('white') 

# Create bars
bars1 = ax.bar(x - width/2, t5_scores, width, label='T5', color=colors[0], alpha=0.9)
bars2 = ax.bar(x + width/2, llama_scores, width, label='LLaMA', color=colors[1], alpha=0.9)

# Customize appearance
ax.set_ylabel('Scores', fontsize=15)

ax.set_xticks(x)
ax.set_xticklabels(metrics, fontsize=15)
ax.set_ylim(-1.2, 1.2)  # Extended y-axis limits to provide more padding

# Add grid for better readability
ax.yaxis.grid(True, linestyle='--', alpha=0.3)
ax.set_axisbelow(True)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)



# Annotate bars with improved positioning and formatting
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}', 
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3 if height >= 0 else -14),
                    textcoords="offset points",
                    ha='center', va='bottom' if height >= 0 else 'top',
                    fontsize=13,
                    )

plt.tight_layout()
plt.savefig('plot500.png', dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
